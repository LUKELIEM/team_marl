## Tribes and Agents

### Tribal Attachment
Under tribal organization, every agent attaches itself to a tribe:
* Adopt a tribal color
* Receive a tribal reward on top of reward doled out by the environment
* Is able to tell if the other agents in its view box are "US" (same tribe) or "THEM" (not the same tribe)

### Observation
The environment still receives actions from all the agents, determines the next state, and doles out new observations and rewards to the agents. While the reward to each agent remains the same, the observation will contain more information:

* location of food unit (as before)
* location of wall (as before)
* location of "US" agents in view box
* location of "Them" agents in view box

In the future, when we implement specialization within a tribe, each agent will belong to a caste, such as gatherer, warrior, scouts, etc. The observation from the environment will need to contain these caste-specific information. 

An additional enhancement will be have the laser tag zone be narrower and shorter than the view box. The observation from the environment will need to contain these additional information.

### Tribe Class
A tribal entity (Tribe Class) need to perform these functions:
* Set tribal color
* Accept agent attachment
* Process tribe-specific information from the environment that is relevant to doling out tribal reward, such as how many "US" agents and "THEM" agents have been tagged when one of its agents fires its laser.
* Assign tribal reward to each tribal agent
* Maintain tribe-specific statistics

## Tribe Class

The tribe, as an imaginary entity, is assigned a tribe name and a color.

Later on, it will perform the important function of:

- doling out non-environmental tribal reward/bonus to incent or penalize agent behaviors


```python
import numpy as np

"""
In this file, we will implement the Tribe Class:
- Agents attached to the Tribe have a "Us" versus "Them" mentality.
- The Tribe awards agents with tribal reward based on Culture.
"""
class Tribe():
    
    color = None
    name = None
    culture = None
    members = []

    # This is run when the tribe is created
    def __init__(self, name='Vikings', color='red', culture={'name':'individualist'}, agents=[]):
        self.color = color
        self.name = name
        self.culture = culture
        self.members = agents
        
        for agent in self.members:
            agent.attach_to_tribe(name, color)
    
    def sum_rewards(self):
        # Add the rewards of the tribe members elementwise
        rewards = [0 for i in range(len(self.members[0].rewards))]
        for a in self.members:
            rewards = [rewards[j] + a.rewards[j] for j in range(len(rewards))]
        return rewards   # return the tribe's rewards 
    
    def sum_US_hits(self):
        # Add the num of friendly fires (fragging) by the tribe members elementwise
        US_hits = [0 for i in range(len(self.members[0].US_hits))]
        for a in self.members:
            US_hits = [US_hits[j] + a.US_hits[j] for j in range(len(US_hits))]
        return US_hits   # return the tribe's fragging incidents

    
    def tribal_awards(self, US_hits = None, THEM_hits = None, tag_hist=None):
        culture = self.culture['name']
        if culture is 'cooperative':
            coop_factor = self.culture['coop_factor']
            rewards_to_tribe = self.sum_rewards() 
            num_members = len(self.members)
            awards = [r/num_members*coop_factor for r in rewards_to_tribe]
        elif culture is 'individualist':
            rewards_to_tribe = self.sum_rewards() 
            awards = [0 for r in rewards_to_tribe]
        elif culture is 'no_fragging':
            penalty = self.culture['penalty']
            awards = [friendly_fire * penalty for friendly_fire in US_hits]
        elif culture is 'pacifist':
            penalty = self.culture['penalty']
            awards = [laser * penalty for laser in tag_hist]
        elif culture is 'warlike':
            penalty = self.culture['penalty']
            reward = self.culture['reward']
            awards = [friendly_fire * penalty + enemy_hit * reward \
                      for friendly_fire, enemy_hit in zip(US_hits,THEM_hits)]
        else:
            awards = None
        return awards
```

## Multiple agent - Cooperate Incentive

An agent should cooperate with other agents of its own tribe. 

**Tribal Reward** to encourage maximizing total tribal reward:
agent_reward = agent_reward + coop  
coop = min(agent_reward x coop_rate, team_reward/num_learners x 0.2)  

If coop_rate is 50%, the agent is responsible fo the environmental reward he earns, while how the team does as a whole can boost its reward by a further 50%.
